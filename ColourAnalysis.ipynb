```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the CSV file into a DataFrame
df = pd.read_csv('car_prices.csv')

# Remove rows where 'color' column contains numerical values, 'nan', '-', or '—'
valid_colors = df['color'].astype(str).str.strip().str.lower().isin([
    'black', 'white', 'silver', 'gray', 'blue', 'red', 'green', 'gold',
    'beige', 'burgundy', 'brown', 'orange', 'purple', 'off-white', 'yellow',
    'charcoal', 'turquoise', 'pink', 'lime'
])
cleaned_df = df[valid_colors]

# Get the count of each color in the cleaned DataFrame
color_counts = cleaned_df['color'].value_counts()

# Create a color map dictionary for color-coded pie chart
color_map = {'black': '#000000', 'white': '#FFFFFF', 'silver': '#C0C0C0', 'gray': '#808080',
             'blue': '#0000FF', 'red': '#FF0000', 'green': '#008000', 'gold': '#FFD700',
             'beige': '#F5F5DC', 'burgundy': '#800020', 'brown': '#A52A2A', 'orange': '#FFA500',
             'purple': '#800080', 'off-white': '#F5F5F5', 'yellow': '#FFFF00', 'charcoal': '#36454F',
             'turquoise': '#40E0D0', 'pink': '#FFC0CB', 'lime': '#00FF00'}

# Create a pie chart with color-coded wedges and remove text inside the pie chart
plt.figure(figsize=(8, 8))
wedges, texts, _ = plt.pie(color_counts, labels=None, colors=[color_map[color] for color in color_counts.index], startangle=140, autopct='%1.1f%%')

# Create legend with color-coded patches
plt.legend(wedges, color_counts.index, loc='center left', bbox_to_anchor=(1, 0.5))

plt.title('Distribution of Car Colors')

plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle

plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the CSV file into a DataFrame
df = pd.read_csv('car_prices.csv')

# Remove rows where 'color' column contains numerical values, 'nan', '-', or '—'
valid_colors = df['color'].astype(str).str.strip().str.lower().isin([
    'black', 'white', 'silver', 'gray', 'blue', 'red', 'green', 'gold',
    'beige', 'burgundy', 'brown', 'orange', 'purple', 'off-white', 'yellow',
    'charcoal', 'turquoise', 'pink', 'lime'
])
cleaned_df = df[valid_colors]

# Get the count of each color in the cleaned DataFrame
color_counts = cleaned_df['color'].value_counts()

# Define color names and their respective hex codes
color_map = {'black': '#000000', 'white': '#FFFFFF', 'silver': '#C0C0C0', 'gray': '#808080',
             'blue': '#0000FF', 'red': '#FF0000', 'green': '#008000', 'gold': '#FFD700',
             'beige': '#F5F5DC', 'burgundy': '#800020', 'brown': '#A52A2A', 'orange': '#FFA500',
             'purple': '#800080', 'off-white': '#F5F5F5', 'yellow': '#FFFF00', 'charcoal': '#36454F',
             'turquoise': '#40E0D0', 'pink': '#FFC0CB', 'lime': '#00FF00'}

# Create a bar graph with color-coded bars
plt.figure(figsize=(10, 6))
bars = plt.bar(color_counts.index, color_counts.values, color=[color_map[color] for color in color_counts.index])

# Add legend with color-coded patches
legend_handles = [plt.Rectangle((0,0),1,1, color=color_map[color]) for color in color_counts.index]
plt.legend(legend_handles, color_counts.index, loc='center left', bbox_to_anchor=(1, 0.5))

plt.title('Distribution of Car Colors')
plt.xlabel('Color')
plt.ylabel('Count')
plt.xticks(rotation=45, ha='right')  # Rotate x-axis labels for better readability
plt.tight_layout()  # Adjust layout to prevent overlap of labels

plt.show()
```
